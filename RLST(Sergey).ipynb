```python
# Словарь, содержащий имена столбцов и их замещающие значения.
col_replacements = {'Alley': 'NoAccess',
                    'PoolQC': 'NoPool',
                    'Fence': 'NoFence',
                    'MiscFeature': 'NoMisc',
                    'FireplaceQu': 'NoFP',
                    'GarageType': 'NoGarage',
                    'GarageFinish': 'NoGarage',
                    'GarageQual': 'NoGarage',
                    'GarageCond': 'NoGarage',
                    'BsmtQual': 'NoBsmt',
                    'BsmtCond': 'NoBsmt',
                    'BsmtExposure': 'NoBsmt',
                    'BsmtFinType1': 'NoBsmt',
                    'BsmtFinType2': 'NoBsmt'}

# заменить значения NaN в каждом столбце, используя словарь
df.replace(np.nan, col_replacements, regex=True, inplace=True)
```

```python
# Замена пропущенных значений в столбце 'MasVnrType' на строку 'None'
df['MasVnrType'].fillna('None', inplace=True)

# Вывод количества каждого типа кладки в столбце 'MasVnrType', включая пропущенные значения
df['MasVnrType'].value_counts(dropna=False)
```

```python
# Замена пропущенных значений в столбце 'Electrical' на строку 'Mix'
df['Electrical'].fillna('Mix', inplace=True)

# Вывод количества каждого типа электрических систем в столбце 'Electrical', включая пропущенные значения
df['Electrical'].value_counts(dropna=False)
```

```python
# Заполнение пропущенных значений в столбце 'LotFrontage' средним значением
df['LotFrontage'].fillna(df['LotFrontage'].mean(), inplace=True)

# Вывод количества каждого значения в столбце 'LotFrontage', включая пропущенные значения
df['LotFrontage'].value_counts(dropna=False)
```

```python
# Заполнение пропущенных значений в столбце 'MasVnrArea' нулем
df['MasVnrArea'].fillna(0, inplace=True)

# Вывод количества каждого значения в столбце 'MasVnrArea', включая пропущенные значения
df['MasVnrArea'].value_counts(dropna=False)
```

```python
# Заполнение пропущенных значений в столбце 'GarageYrBlt' средним значением
df['GarageYrBlt'].fillna(df['GarageYrBlt'].mean(), inplace=True)

# Вывод количества каждого значения в столбце 'GarageYrBlt', включая пропущенные значения
df['GarageYrBlt'].value_counts(dropna=False)
```

```python
# Импортируем класс OneHotEncoder из библиотеки sklearn.preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

# Создаем экземпляр класса OneHotEncoder
ohe = OneHotEncoder()

# Применяем метод fit_transform к датафрейму df и сохраняем полученные значения в переменную encoded.
encoded = ohe.fit_transform(df)

ss = StandardScaler()
ss_df = ss.fit_transform(encoded.toarray())

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Создаем массив X из категориальных признаков
X = [['a', 'b'], ['b', 'c'], ['c', 'a']]

# Применяем метод fit_transform() экземпляра OneHotEncoder к массиву X
encoded_X = ohe.fit_transform(X)

# Масштабируем данные с помощью StandardScaler
scaler = StandardScaler()
scaled_X = scaler.fit_transform(encoded_X.toarray())

# Создаем массив y с метками классов
y = [0, 1, 0]

# Обучаем модель LogisticRegression на масштабированных данных
clf = LogisticRegression()
clf.fit(scaled_X, y)
```
